```python
import shapefile
sf = shapefile.Reader('范围(2014.10).shp')
Shapes = sf.shapes()
```

```python
import matplotlib.pyplot as plt
plt.figure()
for shape in sf.shapeRecords():
    x = [i[0] for i in shape.shape.points[:]]
    y = [i[1] for i in shape.shape.points[:]]
    plt.plot(x,y)
plt.show()
```
